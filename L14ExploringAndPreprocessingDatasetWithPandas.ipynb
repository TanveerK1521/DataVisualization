```python
df_can = pd.read_excel(
    'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx',
    sheet_name='Canada by Citizenship',
    skiprows=range(20),
    skipfooter=2)
df_can.head()
# df.tail for bottom rows
```

```python
# When analyzing a dataset it is good idea to start by getting basic information about the dataframe.
df_can.info(verbose=False)
```

```python
# Now lets clean the data and remove some unnecessary columns
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head()
```

3. METHOD 2, MORE ROBUST and can filter on multiple column names df['column'] # returns series and df[['column1','column2']] # returns dataframe

SELECTING ROWS  
There are two main ways to select rows:  
df.loc[label] # filter by the labels of index/column  
df.iloc[index] # filter by position of index/column  


```python
# setting country column as index column
df_can.set_index('Country',inplace = True)
df_can.head(1)
```

```python
# Option to remove name of index
df_can.index.name = None
df_can.head(1)
```

```python
# Lets view number of immigrants from JAPAN(row 87) for following scenario
# 1. The full row data(all columns)
df_can.loc['Japan']   # or alternatively df_can.iloc[87]
```

```python
# COlumn names that are integers such as years might introduce some problem and confusion.
# To avoid this we convert column names into strings from 1980 to 2013

df_can.columns = list(map(str, df_can.columns))
#[print (type(x)) for x in df_can.columns.values] #<-- uncomment to check type of column headers
```

```python
# Since we converted years to string. lets declare a variable that will allow us to easily call upon full range of years
years =  list(map(str, range(1980, 2014)))
years
```

```python
# Create a list named 'year' using map function for years ranging from 1990 to 2013.
# Then extract the data series from the dataframe df_can for Haiti using year list.
year = list(map(str,range(1990,2014)))
year
```

```python
haiti = df_can.loc['Haiti',year]
haiti
```

FILTERING BASED ON A CRITERIA  
To filter datafram based on a condition, we simply pass the condition as a boolean vector

```python
# 1. create the condition boolean series
condition = df_can['Continent'] == 'Asia'
print(condition)
```

```python
# 2. pass this condition into the dataFrame
df_can[condition]
```

```python
# QUESTION 
#Fetch the data where AreaName is 'Africa' and RegName is 'Southern Africa'.
# Display the dataframe and find out how many instances are there?
df_can[(df_can['Continent']=='Africa') & (df_can['Region']=='Southern Africa')]
```
